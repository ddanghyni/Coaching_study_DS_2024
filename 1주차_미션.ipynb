```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df_new = df[df['Year'] >= 2011]
#df_new = df_new.set_index('Year')
#df_new
```

```python
df_new.pivot_table(index = 'Year',
                   columns = 'Country',
                   values = "Life_Expectancy")
#df_new.pivot(columns = 'Country', values = "Life_Expectancy") index가 year
```

```python
df_new.groupby(['Year', 'Country'])['Life_Expectancy'].mean().unstack()
```
